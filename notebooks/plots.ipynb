# Визуализация результатов моделирования

```python
from collections import defaultdict
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_data(filepath):
    
    with open(filepath, "r", encoding="utf-8") as f:
        lines = f.readlines()
        
    data = defaultdict(list)

    for i, line in enumerate(lines):

        if i == 0:
            continue

        arr = list(map(float, line.split()))

        data["jd"].append(arr[0])
        data["x"].append(arr[1])
        data["y"].append(arr[2])
        data["z"].append(arr[3])
        data["vx"].append(arr[4])
        data["vy"].append(arr[5])
        data["vz"].append(arr[6])
        data["q0"].append(arr[7])
        data["q1"].append(arr[8])
        data["q2"].append(arr[9])
        data["q3"].append(arr[10])
        
    return data
    
```

```python
# 86400c время моделирования
data_with_perturbation = read_data("../results/with_perturbation.txt")
data_without_perturbation = read_data("../results/without_perturbation.txt")
```

```python
plt.figure(figsize=(12, 12))

plt.subplot(3, 2, 1)
plt.plot(data_without_perturbation["x"], data_without_perturbation["y"], linewidth=1)
plt.grid()
plt.title("XY without perturbation", fontsize=12)
plt.axis("square")

plt.subplot(3, 2, 3)
plt.plot(data_without_perturbation["x"], data_without_perturbation["z"], linewidth=1)
plt.grid()
plt.title("XZ without perturbation", fontsize=12)
plt.axis("square")

plt.subplot(3, 2, 5)
plt.plot(data_without_perturbation["y"], data_without_perturbation["z"], linewidth=1)
plt.grid()
plt.title("YZ without perturbation", fontsize=12)
plt.axis("square")

plt.subplot(3, 2, 2)
plt.plot(data_with_perturbation["x"], data_with_perturbation["y"], linewidth=1)
plt.grid()
plt.title("XY with perturbation", fontsize=12)
plt.axis("square")

plt.subplot(3, 2, 4)
plt.plot(data_with_perturbation["x"], data_with_perturbation["z"], linewidth=1)
plt.grid()
plt.title("XZ with perturbation", fontsize=12)
plt.axis("square")

plt.subplot(3, 2, 6)
plt.plot(data_with_perturbation["y"], data_with_perturbation["z"], linewidth=1)
plt.grid()
plt.title("YZ with perturbation", fontsize=12)
plt.axis("square");
```

```python
plt.figure(figsize=(12, 6))

plt.plot(
    data_without_perturbation["jd"], 
    np.array(data_without_perturbation["q0"]) - np.array(data_with_perturbation["q0"]), 
    label="$\Delta_{q0}$",
    color="black",
    linestyle="--",
)

plt.grid()
plt.xlabel("julian day", fontsize=12)
plt.legend(fontsize=12);
```
